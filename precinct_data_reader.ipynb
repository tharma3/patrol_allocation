```python
import pandas as pd
import os
```

```python
os.getcwd()
```

```python
files = os.listdir()
```

```python
files = [file for file in files if file[-3:]=='csv']
names = [file[:-4] for file in files]
```

```python
frames = []
for file, name in zip(files, names):
    name = pd.read_csv(file)
    frames.append(name)
```
