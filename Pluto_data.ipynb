```python
bk = pd.read_csv('PLUTO_for_WEB/BK_18v1.csv')
bx = pd.read_csv('PLUTO_for_WEB/BX_18v1.csv')
mn = pd.read_csv('PLUTO_for_WEB/MN_18v1.csv')
qn = pd.read_csv('PLUTO_for_WEB/QN_18v1.csv')
si = pd.read_csv('PLUTO_for_WEB/SI_18v1.csv')
```

```python
frames = [bk, bx, mn, qn, si]
nycpluto = pd.concat(frames)
```

```python
nycpluto.head()
```

```python
nycpluto.columns
```

```python
nycplutoR = nycpluto[['Address', 'AssessTot', 'BBL', 'BldgArea', 'BldgClass', 'Block', 'BoroCode',
                     'Borough', 'ComArea', 'CT2010', 'LandUse', 'PolicePrct', 'ResArea', 'RetailArea', 'Tract2010',
                     'UnitsRes', 'UnitsTotal', 'XCoord', 'YCoord', 'YearBuilt', 'ZipCode', 'ZoneDist1', 
                      'ZoneDist2', 'ZoneDist3', 'ZoneDist4', 'AssessLand', 'RetailArea']]
```

```python
nycplutoR.describe()
```

```python
# removing some weird values in the year column
nycplutoR = nycplutoR[(nycplutoR['YearBuilt'] != 0) & (nycplutoR['YearBuilt'] < 2020)]
```

```python
nycplutoR.describe()
```

```python
# grouping by police precinct, taking the mean for each measure
plutoG = nycplutoR.groupby(by='PolicePrct').mean().reset_index()
plutoG.head()
```

```python
# reducing to columns of interest
plutoG2 = plutoG[['PolicePrct', 'AssessLand', 'AssessTot', 'ComArea', 'LandUse', 'ResArea', 'RetailArea', 
                  'UnitsRes', 'UnitsTotal', 'YearBuilt']]
```

```python
plutoG2 = plutoG2.astype(int)
```

```python
plutoG2.head()
```

```python
# renaming column titles
plutoG2.rename(columns={'PolicePrct':'precinct', 'AssessTot':'assess_total_mean', 'AssessLand':'assess_land_mean',
                       'LandUse':'land_use_mean', 'ResArea':'res_area_mean', 'RetailArea':'retail_area_mean',
                        'UnitsRes':'units_res_mean', 'UnitsTotal':'units_total_mean', 'YearBuilt':'yearbuilt',
                       'ComArea':'comm_area_mean'}, inplace=True)
```
