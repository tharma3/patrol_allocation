```python
import pandas as pd
import os
```

```python
files = os.listdir()
```

```python
files = [file for file in files if file[-3:]=='csv']
names = [file[:-4] for file in files]
```

```python
frames = []
for file, name in zip(files, names):
    name = pd.read_csv(file)
    frames.append(name)
```

```python
crime = frames[3] #refers to complaint_data.csv
```

```python
sf = frames[2]
sf.head()
```

## merging dfs and adding population

```python
## merge the dataframes
sfCrime = pd.merge(crime, sf, left_on='ADDR_PCT_CD',
                  right_on='precinct')
```

```python
sfCrime.head()
```

## Adding in population and merging with crime df

```python
pop = frames[6]
pop2 = pop[['precinct_me', 'total_pop2017']]
pop2.head()
```

```python
crimeF= pd.merge(sfCrime, pop2, left_on='ADDR_PCT_CD',
                  right_on='precinct_me')
```

```python
crimeF.drop(['precinct_me'], axis=1, inplace=True)
```

```python
crimeF.head()
```

```python
## Creating Rates, Creating Column that represents sums of all rates
```

```python
a = crimeF.values
```

```python
crimeF.iloc[:,1:53] = a[:,1:53] * 100 # multiply by 100 so that when i divide by population 
                                      # later, the value is a percent 
```

```python
crimeF.iloc[:,1:53] = a[:,1:53] / a[:,53,None] # dividing each value by the precinct population 
```

```python
crimeRates = crimeF # renaming for clarity
crimeRates.head()
```

```python
crimeRates.ADDR_PCT_CD.dtype
```

```python
# creating value for total crime rate per precinct
crimeRates['total_crime']= crimeRates.iloc[:, 1:53].sum(axis=1)
```

```python
crimeRates.head()
```

```python
# creating new column as a percentage of total crime in the city, based sum of total_crime column
# each value will represent the share of total crime that occurs in that precinct 
crimeRates['total_crime_share'] = 100*(crimeRates.total_crime / crimeRates.total_crime.sum())
```
