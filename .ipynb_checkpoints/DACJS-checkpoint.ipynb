```python
url = 'https://data.cityofnewyork.us/api/geospatial/v2h8-6mxf?method=export&format=Shapefile'
block = gp.GeoDataFrame.from_file(url)
```

```python
block.to_crs(epsg=2263, inplace=True)
```

```python
precinct_block = gp.sjoin(block, precinct, how='left', op='intersects')
```

```python
precinct_block.fillna(0, inplace=True)
```

```python
no_cinct_b = precinct_block[precinct_block['precinct'] == 0 ]
```

```python
precinct_block['precinct'].unique()
```

```python
precinct_block_copy = precinct_block
```

```python
precinct_block.tail(1)
```

```python
precinct_block = precinct_block_copy
precinct_block = precinct_block[['bctcb2010', 'boro_code', 'cb2010', 'ct2010', 'precinct']]
precinct_block = precinct_block.rename(columns={'precinct':'precinct_me'})

precinct_block.head()
```

```python
precinct_block['cb2010'] = precinct_block['cb2010'].str[:-3]
```

```python
precinct_block['ct2010'] = precinct_block['ct2010'].astype('int')
precinct_block['boro_code'] = precinct_block['boro_code'].astype('int')
precinct_block['cb2010'] = precinct_block['cb2010'].astype('int')
```

```python
pop_df['county'] = pop_df['county'].astype('int')
```

```python
precinct_block['boro_code'] = precinct_block['boro_code'].replace([1,2,3,4,5], [61,5,47,81,85])
```

```python
precinct_block.head()
```

```python
print(pop_df.info())
print(precinct_block.info())
```

```python
precinct_pop = pop_df.merge(precinct_block, how='left', left_on=['county','tract', 'block group'],\
                            right_on=['boro_code', 'ct2010', 'cb2010'])
precinct_pop.head()
```

```python
precinct_pop.shape
```

```python
precinct_pop = precinct_pop.drop_duplicates(subset=['county', 'tract', 'block group'])
```

```python
precinct_pop.head()
```

```python
precinct_pop_grouped = precinct_pop.groupby('precinct_me').sum()
```

```python
precinct_pop_grouped.columns
```

```python
precinct_pop_m = precinct_pop_grouped[['total_pop2014', 'total_pop2015',
       'total_pop2016', 'total_pop2017']]
```

```python
total_14 = precinct_pop_m['total_pop2014'].sum()
total_15 = precinct_pop_m['total_pop2015'].sum()
total_16 = precinct_pop_m['total_pop2016'].sum()
total_17 = precinct_pop_m['total_pop2017'].sum()

cops = 20000

precinct_pop_m['model_14'] = (precinct_pop_m['total_pop2014']/total_14) * cops
precinct_pop_m['model_15'] = (precinct_pop_m['total_pop2015']/total_15) * cops
precinct_pop_m['model_16'] = (precinct_pop_m['total_pop2016']/total_16) * cops
precinct_pop_m['model_17'] = (precinct_pop_m['total_pop2017']/total_17) * cops
```

```python
precinct_pop_m = precinct_pop_m.reindex(axis=1)
```

```python
precinct_pop_m.to_csv('precinct_pop_m.csv')
```
