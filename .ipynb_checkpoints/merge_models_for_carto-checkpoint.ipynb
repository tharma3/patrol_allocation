## RACE AND INCOME LAYER

## NO ARREST RATES, QUANTILES, and QOL ARRESTS PER CAPITA

## Bring it all together!

```python
all_data = overpoliced.merge(race_and_income\
            [['precinct','percent_ppl_of_color','median_fam_income']], 
            on='precinct').merge\
        (pop_and_crime[['ADDR_PCT_CD','allocation']], 
         left_on='precinct',right_on='ADDR_PCT_CD').rename(columns=\
        {'allocation':'pop_and_crime_po'})\
        .merge(pluto_acs[['precinct','allocation']], on='precinct')\
        .rename(columns={'allocation':'pluto_acs_po'})\
        .merge(crashes[['precinct','officers']], on='precinct')\
        .rename(columns={'officers':'crashes_po'})\
        .merge(arrests_violations[['ARREST_PRECINCT','officers']], 
        left_on='precinct',right_on='ARREST_PRECINCT')\
        .rename(columns={'officers':'arrest_viola_po'})\
        .merge(cfs_vio[['Precinct','officers']],
        left_on='precinct', right_on='Precinct')\
        .rename(columns={'officers':'cfs_po'})\
        .merge(violent[['ARREST_PRECINCT','MODEL_allocation_14']], 
        left_on='precinct', right_on='ARREST_PRECINCT')\
        .rename(columns={'MODEL_allocation_14':'violent_po'})\
        .merge(gentrification[['precinct','MODEL_allocation']],on='precinct')\
        .rename(columns={'MODEL_allocation':'gentrification_po'})\
        .merge(day_pop[['precinct','MODEL_allocation']], on='precinct')\
        .rename(columns={'MODEL_allocation':'adjusted_day_pop_po'})
                                              

    
    
    
    
all_data.head()                                              
                                              
```

```python
all_data.plot()
```
